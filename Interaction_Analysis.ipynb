```python
#load data
indir = '/content/drive/MyDrive/Capstone/dataset/'
PATH_IN = os.path.join(indir, 'goodreads_interactions.csv')
df_interactions = pd.read_csv(
    PATH_IN,
    usecols=["user_id", "book_id", "rating"]
)
```

```python
print("Total users:", df_interactions['user_id'].nunique())
print("Total books:", df_interactions['book_id'].nunique())
print("Total ratings (non-missing):", df_interactions['rating'].notna().sum())
avg_ratings_per_book = df_interactions.shape[0] / df_interactions['book_id'].nunique()
avg_books_per_user = df_interactions.shape[0] / df_interactions['user_id'].nunique()

print(f"Average number of ratings per book: {avg_ratings_per_book:.2f}")
print(f"Average number of books rated per user: {avg_books_per_user:.2f}")
```
