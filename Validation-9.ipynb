```python



7777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777777#!/usr/bin/env python
# validation_userRecall_v4.py   · 2025-05-23  (with tqdm)
# -------------------------------------------------------------------
# • k fixed at 10
# • Recall = hits / (sampled_users × 10)
# • Sampled users per anchor = min(MAX_USERS, ceil(total_readers * RATIO))
# • Outputs three metric sets: SBERT / Collab / Fused (deduped, padded to 10)
# • tqdm progress bar visualizes reading & evaluation
# -------------------------------------------------------------------

import os, gzip, json, math, argparse, random
from collections import defaultdict
import numpy as np, pandas as pd
from tqdm.auto import tqdm   # progress bar

# ================== ★★ EDIT THIS LINE ONLY ★★ ==================
BASE = "/Users/brucewei/Desktop/Capstone"  # ← replace with your Capstone root
# ===============================================================

# ---------- File paths ----------
JSON_PLAIN = os.path.join(BASE, "goodreads_interactions_comics_graphic.json")
JSON_GZIP  = JSON_PLAIN + ".gz"
DEFAULT_JSON = JSON_PLAIN if os.path.exists(JSON_PLAIN) else JSON_GZIP

SBERT_CSV  = os.path.join(BASE, "output_batches", "recommender_all_dedup.csv")
COLLAB_CSV = os.path.join(BASE, "recommendercollab.csv")

# ---------- Evaluation parameters ----------
K         = 10        # Top-K
RATIO     = 0.10      # sampling ratio
MAX_USERS = 50        # per-anchor cap
SEED      = 2025
random.seed(SEED); np.random.seed(SEED)

# ═══════════════ Utility functions ═══════════════
def read_rec(path:str):
    """Load a recommendation CSV (two columns: input_book_id, recommended_book_id)."""
    if not os.path.isfile(path):
        print(f"[WARN] Recommendation file not found → {path}")
        return None
    return pd.read_csv(path, dtype=str,
                       usecols=["input_book_id","recommended_book_id"])

def open_any(p:str):
    """Open .json or .json.gz transparently."""
    return gzip.open(p,"rt",encoding="utf-8") if p.endswith(".gz") \
           else open(p,"r",encoding="utf-8")

def load_inter(path:str, anchors:set[str]):
    """Return book2users and user2books, with tqdm progress."""
    book2users, user2books = defaultdict(list), defaultdict(set)
    with open_any(path) as f:
        for line in tqdm(f, desc="Reading interactions"):
            r = json.loads(line); b,u = r["book_id"], r["user_id"]
            if b in anchors:
                book2users[b].append(u)
            user2books[u].add(b)
    book2users = {b: np.array(u) for b,u in book2users.items()}
    return book2users, user2books

def sample_users(arr:np.ndarray):
    """Randomly sample users under MAX_USERS × RATIO rule (no replacement)."""
    want = min(MAX_USERS, math.ceil(len(arr) * RATIO))
    if want >= len(arr):
        return arr
    return np.random.choice(arr, want, replace=False)

def evaluate(df, anchors, b2u, u2b):
    """Compute macro-precision, recall, and F1."""
    hits = preds = denom = 0
    for a in tqdm(anchors, desc="Evaluating"):
        users_full = b2u.get(a, np.array([]))
        if users_full.size == 0:
            continue
        users = sample_users(users_full)

        truth = set()
        for u in users:
            truth.update(u2b[u] - {a})

        recs = df.loc[df.input_book_id==a,
                      "recommended_book_id"].tolist()[:K]
        h = len(set(recs) & truth)

        hits  += h
        preds += K
        denom += len(users) * K

    P = hits / preds  if preds else 0
    R = hits / denom  if denom else 0
    F = 0 if P+R==0 else 2*P*R/(P+R)
    return P,R,F

def fuse_top10(df1, df2, anchors):
    """Merge two lists, deduplicate, keep top-10 per anchor."""
    rows=[]
    for a in anchors:
        cand = list(df1[df1.input_book_id==a].recommended_book_id[:15]) + \
               list(df2[df2.input_book_id==a].recommended_book_id[:15])
        uniq, seen = [], set()
        for b in cand:
            if b not in seen:
                seen.add(b); uniq.append(b)
            if len(uniq)==10: break
        rows.extend([(a,b) for b in uniq])
    return pd.DataFrame(rows,
                        columns=["input_book_id","recommended_book_id"])

# ═══════════════ Main ═══════════════
def main(json_path):
    sb = read_rec(SBERT_CSV)
    co = read_rec(COLLAB_CSV)
    dfs = [d for d in (sb,co) if d is not None]
    if not dfs:
        raise SystemExit("No recommendation CSV found; check the paths!")

    anchors = pd.concat([d.input_book_id for d in dfs]).unique()
    print(f"[INFO] Anchors: {len(anchors):,}")

    b2u,u2b = load_inter(json_path, set(anchors))
    print(f"[INFO] Interactions loaded (anchors with readers = {len(b2u):,})")
    print(f"[INFO] Sampling ratio {RATIO*100:.0f}%, MAX_USERS per anchor {MAX_USERS}")

    if sb is not None:
        P,R,F = evaluate(sb, anchors, b2u, u2b)
        print(f"SBERT   P@10={P:.4f}  R@10={R:.4f}  F1={F:.4f}")

    if co is not None:
        P,R,F = evaluate(co, anchors, b2u, u2b)
        print(f"Collab  P@10={P:.4f}  R@10={R:.4f}  F1={F:.4f}")

    if sb is not None and co is not None:
        fused = fuse_top10(sb, co, anchors)
        P,R,F = evaluate(fused, anchors, b2u, u2b)
        print(f"Fused   P@10={P:.4f}  R@10={R:.4f}  F1={F:.4f}")

# ═══════════════ Entry point ═══════════════
if __name__ == "__main__":
    parser = argparse.ArgumentParser("Validation (tqdm, k=10, user×k recall)")
    parser.add_argument("--json", default=DEFAULT_JSON,
                        help="Interaction file (.json or .json.gz)")

    # ★★ use parse_known_args() to ignore Jupyter's --f ★★
    args = parser.parse_known_args()[0]

    if BASE.startswith("/absolute/path"):
        raise SystemExit(" Please set BASE to your absolute Capstone path at the top of this script.")

    if not os.path.isfile(args.json):
        raise SystemExit("Interaction file not found; check the path or use --json to specify it.")
    main(args.json)
```
